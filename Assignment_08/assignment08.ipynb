```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("census_costs.txt", skiprows=1, sep="\t")
df.head(3)
```

```python
def cleanse_py(a_string):
    return int(a_string.replace(",","").replace("*",""))
```

```python
df["year"] = df["Census Year"].apply(cleanse_py)  
df["tot_pop"] = df["Total Population"].apply(cleanse_py)
df
```

```python
def cleanse_cost(b_string):
    return (b_string.replace(",","").replace("$","").replace(" Billion","000000000"))
  
```

```python
df["census_cst"] = df["Census Cost"].apply(cleanse_cost)
df["census_cst"] = pd.to_numeric(df["census_cst"])
df
```

```python
def cleanse_Average(c_string):
    return(c_string.replace(",","").replace("$","").replace("cents",""))
```

```python
df["avg_cost"] = df["Average Cost Per Person"].apply(cleanse_Average)
df["avg_cost"] = pd.to_numeric(df["avg_cost"])
df
```

```python
df.describe()
```

```python
fig = px.bar(df, y='avg_cost', x='year', width = 1000 , height = 500)
fig.show()
```

```python
fig = px.scatter(df,
    x="census_cst",
    y="tot_pop",  
    template="plotly_dark"
)
fig.update_layout(showlegend=False)
fig.show()
```
